```python
# Calculate the total number of flights from Origin to Destination within each Year

total_flights = pd.DataFrame(df_flights.groupby(['Year', 'Origin', 'Dest']).size().reset_index(name='Total Flights'))

total_flights
```

```python
# Print the output in the desired format

for index, row in total_flights.iterrows():
    print(f"The total number of flights with year {row['Year']} from origin {row['Origin']} to {row['Dest']} with total number {row['Total Flights']}")
```

```python
# Filter data for flights originating from Mumbai
mumbai_flights = total_flights[total_flights['Origin'] == 'MUMBAI']

# Group by Year and Destination and calculate total flights
result = mumbai_flights.groupby(['Year', 'Dest'])['Total Flights'].sum().reset_index()

# Print the result in sentence format
for _, row in result.iterrows():
    year = row['Year']
    dest = row['Dest']
    total_flights = row['Total Flights']
    print(f"In {year}, there were {total_flights} flights from MUMBAI to {dest}.")
```

```python
# Group by Destination and calculate total flights
total_flights_by_city = mumbai_flights.groupby('Dest')['Total Flights'].sum().reset_index()

# Print the result in sentence format
for _, row in total_flights_by_city.iterrows():
    dest = row['Dest']
    total_flights = row['Total Flights']
    print(f"The total number of flights from Mumbai to {dest} over all the years is {total_flights}.")
```
