### Checking/ Analyzing data related to passengers i.e. features related to Pax(from origin and to origin)

**Pax From Origin:** This term refers to the number of passengers departing from the origin airport.

**Pax To Origin:** This term refers to the number of passengers arriving at the origin airport from other locations.


```python

df_flights[['Pax From Origin']]
```

```python
df_flights[['Pax To Origin']]
```

```python
df_flights['Pax From Origin'].sum(), df_flights['Pax To Origin'].sum()
```

```python
for year in df_flights['Year'].unique():
    pax_from_origin = df_flights[df_flights['Year'] == year]['Pax From Origin'].sum()
    pax_to_origin = df_flights[df_flights['Year'] == year]['Pax To Origin'].sum()
    print(f"In {year}, the total number of passengers from Mumbai is {pax_from_origin} and to Mumbai is {pax_to_origin}.")
```

```python
previous_year_pax_from_origin = None
previous_year_pax_to_origin = None

for year in df_flights['Year'].unique():
    pax_from_origin = df_flights[df_flights['Year'] == year]['Pax From Origin'].sum()
    pax_to_origin = df_flights[df_flights['Year'] == year]['Pax To Origin'].sum()
    
    if previous_year_pax_from_origin is not None and previous_year_pax_to_origin is not None:
        percent_change_from_origin = ((pax_from_origin - previous_year_pax_from_origin) / previous_year_pax_from_origin) * 100
        percent_change_to_origin = ((pax_to_origin - previous_year_pax_to_origin) / previous_year_pax_to_origin) * 100
        print(f"In {year}, the total number of passengers from Mumbai is {pax_from_origin} ({percent_change_from_origin:.2f}% change from previous year) and to Mumbai is {pax_to_origin} ({percent_change_to_origin:.2f}% change from previous year).")
    else:
        print(f"In {year}, the total number of passengers from Mumbai is {pax_from_origin} and to Mumbai is {pax_to_origin}.")
    
    previous_year_pax_from_origin = pax_from_origin
    previous_year_pax_to_origin = pax_to_origin
```

```python
# Calculate the total amount of freight from origin for each year
total_freight_from_origin_per_year = df_flights.groupby('Year')['Freight From Origin'].sum().reset_index()

# Find the year with the maximum and minimum amount of freight from origin
max_freight_from_origin_year = total_freight_from_origin_per_year.loc[total_freight_from_origin_per_year['Freight From Origin'].idxmax()]
min_freight_from_origin_year = total_freight_from_origin_per_year.loc[total_freight_from_origin_per_year['Freight From Origin'].idxmin()]

# Calculate the difference between the maximum and minimum amount of freight from origin
freight_from_origin_difference = max_freight_from_origin_year['Freight From Origin'] - min_freight_from_origin_year['Freight From Origin']

# Print the results
print(f"Year with maximum amount of freight from origin is {int(max_freight_from_origin_year['Year'])} with {max_freight_from_origin_year['Freight From Origin']} kilograms.")
print(f"Year with minimum amount of freight from origin is {int(min_freight_from_origin_year['Year'])} with {min_freight_from_origin_year['Freight From Origin']} kilograms.")
print(f"Difference between maximum and minimum amount of freight from origin is {freight_from_origin_difference} kilograms")
```

```python
# Calculate the total number of passengers from and to Mumbai for each year
pax_data = []
for year in df_flights['Year'].unique():
    pax_from_origin = df_flights[df_flights['Year'] == year]['Pax From Origin'].sum()
    pax_to_origin = df_flights[df_flights['Year'] == year]['Pax To Origin'].sum()
    pax_data.append((year, pax_from_origin, pax_to_origin))

# Sort the data by the sum of passengers from and to Mumbai in descending order
pax_data.sort(reverse=True, key=lambda x: x[1] + x[2])

# Print the results
for year, pax_from_origin, pax_to_origin in pax_data:
    print(f"In {year}, the total number of passengers from Mumbai is {pax_from_origin} and to Mumbai is {pax_to_origin}.")
```

```python
# Calculate the percentile increase or decrease in the amount of freight from and to Mumbai with respect to the previous year
previous_year_freight_from_origin = None
previous_year_freight_to_origin = None

for year in df_flights['Year'].unique():
    freight_from_origin = df_flights[df_flights['Year'] == year]['Freight From Origin'].sum()
    freight_to_origin = df_flights[df_flights['Year'] == year]['Freight To Origin'].sum()
    
    if previous_year_freight_from_origin is not None and previous_year_freight_to_origin is not None:
        percent_change_from_origin = ((freight_from_origin - previous_year_freight_from_origin) / previous_year_freight_from_origin) * 100
        percent_change_to_origin = ((freight_to_origin - previous_year_freight_to_origin) / previous_year_freight_to_origin) * 100
        print(f"In {year}, the total amount of freight from Mumbai is {freight_from_origin.round(2)} ({percent_change_from_origin:.2f}% change from previous year) and to Mumbai is {freight_to_origin.round(2)} ({percent_change_to_origin:.2f}% change from previous year).")
    else:
        print(f"In {year}, the total amount of freight from Mumbai is {freight_from_origin.round(2)} and to Mumbai is {freight_to_origin.round(2)}.")
    
    previous_year_freight_from_origin = freight_from_origin
    previous_year_freight_to_origin = freight_to_origin
```
